## Merging Player Data

```python
import pandas as pd
import sqlite3

# Load CSV files into DataFrames
Time_Per_Game_df = pd.read_csv('Stats/Per_36_min_Stats.csv')
Advanced_Stats_df = pd.read_csv('Stats/Advanced_Stats.csv')
Poss_Per_Game_df = pd.read_csv('Stats/Per_100_Poss_Stats.csv')
Play_by_Play_df = pd.read_csv('Stats/Play_by_Play.csv')
Player_Salary_df = pd.read_csv('Stats/Player_Salary.csv')
Team_Rating_df = pd.read_csv('Stats/Team_Ratings.csv')

# Create an SQLite in-memory database
conn = sqlite3.connect(':memory:')

# Load DataFrames into the database as tables
Time_Per_Game_df.to_sql('Per36Stats', conn, index=False, if_exists='replace')
Advanced_Stats_df.to_sql('AdvancedStats', conn, index=False, if_exists='replace')
Poss_Per_Game_df.to_sql('Per100PossStats', conn, index=False, if_exists='replace')
Play_by_Play_df.to_sql('PlayByPlay', conn, index=False, if_exists='replace')
Player_Salary_df.to_sql('PlayerSalary', conn, index=False, if_exists='replace')
Team_Rating_df.to_sql('TeamRating', conn, index=False, if_exists='replace')


# SQL query to merge all tables on 'Player'
sql_query = """
    SELECT Per36Stats.*, AdvancedStats.*, Per100PossStats.*, PlayByPlay.*, PlayerSalary.*, TeamRating.*
    FROM Per36Stats
    INNER JOIN AdvancedStats ON Per36Stats.Player = AdvancedStats.Player
    INNER JOIN Per100PossStats ON Per36Stats.Player = Per100PossStats.Player
    INNER JOIN PlayByPlay ON Per36Stats.Player = PlayByPlay.Player
    INNER JOIN PlayerSalary ON Per36Stats.Player = PlayerSalary.Player
    INNER JOIN TeamRating ON Per36Stats.Player 
    AND Per36Stats.Age < 35
    AND Per36Stats.Age > 20
    AND PlayerSalary.[2027-28] IS NULL
    AND PlayerSalary.[2028-29] IS NULL
    AND PlayerSalary.[2029-30] IS NULL
    AND PlayerSalary.[2026-27] IS NULL
    GROUP BY Per36Stats.Player
    """

# Execute the query and load results into a DataFrame
Player_Numbers_df = pd.read_sql_query(sql_query, conn)

# Step 1: Fix Duplicate Column Names

# Save the merged DataFrame to a new CSV file
csv_filename = "Stats/All_Player_Numbers.csv"
Player_Numbers_df.to_csv(csv_filename, index=False)

# Close the connection
conn.close()

# Provide the file to the user
Player_Numbers_df

```

## **Inputs: Features Considered for Trade Suggestions**
| **Category**                    | **Metrics** |
|---------------------------------|------------|
| **A. Player Performance Metrics** | **Scoring**: FG%, 3P%, 2P%, eFG%, TS%, PTS <br> **Playmaking**: AST, AST%, TOV, TOV% <br> **Rebounding**: ORB, DRB, TRB, ORB%, DRB% <br> **Defense**: STL, BLK, STL%, BLK%, DBPM <br> **Usage Rate & Impact**: USG%, BPM, WS/48, VORP |
| **B. Team Fit & Trade Value**   | **Offensive/Defensive Metrics**: ORtg, DRtg, NRtg, On-Off <br> **Trade Fairness Indicators**: <br> - Change in BPM, WS before/after trade <br> - Net Rating Impact (`NRtg/A`) <br> - Positional Fit (PG%, SG%, SF%, PF%, C%) |
| **C. Salary & Contract Data**   | **Salary Considerations**: <br> - Current and future salary (`2024-25`, `2025-26`, `Guaranteed`) <br> - Trade flexibility (expiring contracts, cap impact) |
| **D. Playstyle & Clustering Attributes** | **Turnover Tendencies**: `BadPass`, `LostBall` <br> **Playstyle Variations**: `OnCourt`, `On-Off`, `DWS`, `OWS` |


```python
import pandas as pd
import sqlite3

# Load CSV file into a DataFrame
Player_Numbers_df = pd.read_csv('Stats/All_Player_Numbers.csv')
Team_Rating_df = pd.read_csv('Stats/Team_Ratings.csv')


# Create an SQLite in-memory database
conn = sqlite3.connect(':memory:')

# Load DataFrame into the database as a table
Player_Numbers_df.to_sql('PlayerNumbers', conn, index=False, if_exists='replace')
Team_Rating_df.to_sql('TeamRating', conn, index=False, if_exists='replace')


# SQL query to extract features for Team Fit & Trade Value
player_performance_query = """
SELECT 
    Player,
    "FG%", "3P%", "2P%", "eFG%", "TS%",
    PTS, AST, "AST%",
    ORB, DRB, TRB, "ORB%", "DRB%",
    STL, BLK, "STL%", "BLK%", DBPM,
    TOV, "TOV%",
    "USG%", BPM, "WS/48", VORP
FROM 
PlayerNumbers
"""

# SQL query to extract features for Team Fit & Trade Value
team_fit_query = """
SELECT 
    Player,
    TR.ORtg, TR.DRtg, TR.NRtg, "On-Off",
    WS, BPM, "NRtg/A",
    "PG%", "SG%", "SF%", "PF%", "C%",
    "3PAr", "FTr"
FROM PlayerNumbers
JOIN TeamRating AS TR ON PlayerNumbers.Team = TR.Team
GROUP BY PlayerNumbers.Player
"""

# SQL query to extract features for Salary & Contract
salary_contract_query = """
SELECT 
    Player,
    "2024-25", "2025-26", Guaranteed
FROM PlayerNumbers
GROUP BY PlayerNumbers.Player
"""

# SQL query to extract features for Playstyle & Clustering
playstyle_query = """
SELECT 
    Player,
    BadPass, LostBall,
    OnCourt, "On-Off", DWS, OWS
FROM PlayerNumbers
GROUP BY PlayerNumbers.Player
"""

# Execute the queries and store the results in DataFrames
Player_Performance_df = pd.read_sql_query(player_performance_query, conn)
Team_Fit_Value_df = pd.read_sql_query(team_fit_query, conn)
Player_Salary_Contract_df = pd.read_sql_query(salary_contract_query, conn)
Playstyle_Clustering_df = pd.read_sql_query(playstyle_query, conn)

# Close the connection
conn.close()

# Save the extracted data to CSV files
Player_Performance_csv = "Stats/INPUT_Player_Performance.csv"
Team_Fit_Value_csv = "Stats/INPUT_Team_Fit_Value.csv"
Player_Salary_Contract_csv = "Stats/INPUT_Player_Salary_Contract.csv"
Playstyle_Clustering_csv = "Stats/INPUT_Playstyle_Clustering_Value.csv"

Player_Performance_df.to_csv(Player_Performance_csv, index=False)
Team_Fit_Value_df.to_csv(Team_Fit_Value_csv, index=False)
Player_Salary_Contract_df.to_csv(Player_Salary_Contract_csv, index=False)
Playstyle_Clustering_df.to_csv(Playstyle_Clustering_csv, index=False)
```

## 1. **Determining the Number of Clusters needed**
1. Elbow Method (Inertia)
2. Silhouette Score (SS)
3. Calinski-Harabasz Index (CHI)
4. Davies-Bouldin Index (DBI)
5. Gap Statistic: Compares WCSS to a null reference distribution.


Based on the provided clustering evaluation metrics:

1. **Calinski-Harabasz Index:** The highest score appears at K=2, but a strong secondary peak at K=7 suggests well-separated clusters.  

2. **Silhouette Score:** The highest silhouette score is at K=2, but K=7 also has a strong peak, indicating well-defined clusters.  

3. **Elbow Method:** The inertia curve drops sharply until K=6-7, where it slows, suggesting K=7 balances compact clusters and minimal within-cluster variance.  

4. **Davies-Bouldin Index:** The lowest index value appears around K=6-7, indicating well-separated, compact clusters, making K=7 an optimal clustering choice.  

Best K=7


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

# Load datasets
Player_Numbers_df

# Define relevant features based on cluster characteristics
features = [
    "FG%", "3P%", "2P%", "eFG%", "TS%", "PTS",  # Scoring
    "AST", "AST%", "TOV", "TOV%", "BadPass", "LostBall",  # Playmaking
    "ORB", "DRB", "TRB", "ORB%", "DRB%",  # Rebounding
    "STL", "BLK", "STL%", "BLK%", "DBPM", "DWS", "OWS", "DRtg", "ORtg",  # Defense
    "USG%", "BPM", "WS/48", "VORP", "OnCourt", "On-Off", "WS",  # Usage
    "PG%", "SG%", "SF%", "PF%", "C%",  # Positional Play
    "3PAr", "FTr"  # Shooting Tendencies
]


# Drop rows with missing values

# Extract player names
players = Player_Numbers_df["Player"]
X = Player_Numbers_df[features]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Determine the optimal number of clusters using the Elbow Method
from sklearn.impute import SimpleImputer

# Create an imputer that replaces NaNs with the mean of the respective column
imputer = SimpleImputer(strategy="mean")  
X_scaled = imputer.fit_transform(X_scaled)

inertia = []
k_range = range(1, 15)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(6, 3))
plt.plot(k_range, inertia, marker="o")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Calculate silhouette scores for different cluster sizes
silhouette_scores = []
k_range = range(2, 15)  

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_scaled)
    silhouette_avg = silhouette_score(X_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot the silhouette scores
plt.figure(figsize=(6, 3))
plt.plot(k_range, silhouette_scores, marker="o", linestyle="-")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for Optimal k")
plt.show()
```

```python
import numpy as np
from sklearn.metrics import calinski_harabasz_score

# Assume X_scaled is the preprocessed dataset

# Calculate Calinski-Harabasz Index for different cluster sizes
chi_scores = []
k_range = range(2, 15)  

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_scaled)
    chi_score = calinski_harabasz_score(X_scaled, cluster_labels)
    chi_scores.append(chi_score)

# Plot the CHI scores
plt.figure(figsize=(6, 3))
plt.plot(k_range, chi_scores, marker="o", linestyle="-")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Calinski-Harabasz Index")
plt.title("Calinski-Harabasz Index for Optimal k")
plt.show()
```

```python
# Calculate Davies-Bouldin Index for different cluster sizes

from sklearn.metrics import davies_bouldin_score

dbi_scores = []
k_range = range(2, 15)  

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_scaled)
    dbi_score = davies_bouldin_score(X_scaled, cluster_labels)
    dbi_scores.append(dbi_score)

# Plot the DBI scores
plt.figure(figsize=(6, 3))
plt.plot(k_range, dbi_scores, marker="o", linestyle="-")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Davies-Bouldin Index")
plt.title("Davies-Bouldin Index for Optimal k")
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Perform Hierarchical Clustering (Agglomerative)
optimal_k = 7  # Same as K-Means for comparison
hc = AgglomerativeClustering(n_clusters=optimal_k, affinity='euclidean', linkage='ward')
cluster_labels_hc = hc.fit_predict(X_scaled)

# Add cluster labels to the dataframe
merged_df["Cluster_HC"] = cluster_labels_hc

# Plot the Dendrogram
plt.figure(figsize=(12, 6))
sch.dendrogram(sch.linkage(X_scaled, method='ward'), truncate_mode='level', p=10)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Players")
plt.ylabel("Distance")
plt.show()
```
