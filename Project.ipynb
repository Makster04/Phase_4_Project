## Merging Player Data

```python
import pandas as pd
import sqlite3

# Load CSV files into DataFrames
Time_Per_Game_df = pd.read_csv('Stats/Per_36_min_Stats.csv')
Advanced_Stats_df = pd.read_csv('Stats/Advanced_Stats.csv')
Poss_Per_Game_df = pd.read_csv('Stats/Per_100_Poss_Stats.csv')
Play_by_Play_df = pd.read_csv('Stats/Play_by_Play.csv')
Player_Salary_df = pd.read_csv('Stats/Player_Salary.csv')
Team_Rating_df = pd.read_csv('Stats/Team_Ratings.csv')
Team_Payroll_df = pd.read_excel('Stats/Team_Payroll.xlsx')


# Create an SQLite in-memory database
conn = sqlite3.connect(':memory:')

# Load DataFrames into the database as tables
Time_Per_Game_df.to_sql('Per36Stats', conn, index=False, if_exists='replace')
Advanced_Stats_df.to_sql('AdvancedStats', conn, index=False, if_exists='replace')
Poss_Per_Game_df.to_sql('Per100PossStats', conn, index=False, if_exists='replace')
Play_by_Play_df.to_sql('PlayByPlay', conn, index=False, if_exists='replace')
Player_Salary_df.to_sql('PlayerSalary', conn, index=False, if_exists='replace')
Team_Rating_df.to_sql('TeamRating', conn, index=False, if_exists='replace')
Team_Payroll_df.to_sql('TeamPayroll', conn, index=False, if_exists='replace')



# SQL query to merge all tables on 'Player'
sql_query = """
SELECT 
    Per36Stats.Player, 
    Per36Stats.Team,

    -- Shooting & Scoring Stats
    MAX(Per36Stats."FG%") AS "FG%", 
    MAX(Per36Stats."3P%") AS "3P%", 
    MAX(Per36Stats."2P%") AS "2P%", 
    MAX(Per36Stats."eFG%") AS "eFG%", 
    MAX(AdvancedStats."TS%") AS "TS%",
    MAX(Per36Stats.PTS) AS "PTS",

    -- Playmaking & Rebounding
    MAX(Per36Stats.AST) AS "AST", 
    MAX(AdvancedStats."AST%") AS "AST%",
    MAX(AdvancedStats."TRB%") AS "TRB%",
    MAX(AdvancedStats."ORB%") AS "ORB%",
    MAX(AdvancedStats."DRB%") AS "DRB%",
    MAX(AdvancedStats."BLK%") AS "BLK%",
    MAX(AdvancedStats."STL%") AS "STL%",
    MAX(AdvancedStats."TRB%") AS "TRB%",
    MAX(Per36Stats.ORB) AS "ORB", 
    MAX(Per36Stats.DRB) AS "DRB", 
    MAX(Per36Stats.TRB) AS "TRB", 
    MAX(AdvancedStats.FTr) AS "FTr",
    MAX(AdvancedStats."3PAr") AS "3PAr",

    -- Defensive Stats
    MAX(Per36Stats.STL) AS "STL", 
    MAX(Per36Stats.BLK) AS "BLK", 
    MAX(Per36Stats.TOV) AS "TOV", 
    MAX(AdvancedStats."TOV%") AS "TOV%",
    MAX(AdvancedStats."USG%") AS "USG%", 

    -- Advanced Metrics
    MAX(AdvancedStats.PER) AS "PER", 
    MAX(AdvancedStats.WS) AS "WS",
    MAX(AdvancedStats."WS/48") AS "WS/48", 
    MAX(AdvancedStats.BPM) AS "BPM", 
    MAX(AdvancedStats.VORP) AS "VORP", 
    MAX(AdvancedStats.OBPM) AS "OBPM", 
    MAX(AdvancedStats.DBPM) AS "DBPM",  
    MAX(AdvancedStats.DWS) AS "DWS", 
    MAX(AdvancedStats.OWS) AS "OWS", 


    -- Play-by-Play & Salary Data
    MAX(PlayByPlay.OnCourt) AS "OnCourt", 
    MAX(PlayByPlay."On-Off") AS "On-Off",  
    MAX(PlayerSalary."2024-25") AS "2024-25",
    MAX(PlayerSalary."Guaranteed") AS "Guaranteed",
    MAX(PlayByPlay.BadPass) AS "BadPass", 
    MAX(PlayByPlay.LostBall) AS "LostBall",
    MAX(PlayByPlay.OnCourt) AS "OnCourt",

    MAX(TeamPayroll."2024-25_Payroll") AS "2024-25_Payroll"

FROM Per36Stats
INNER JOIN AdvancedStats ON Per36Stats.Player = AdvancedStats.Player
INNER JOIN Per100PossStats ON Per36Stats.Player = Per100PossStats.Player
INNER JOIN PlayByPlay ON Per36Stats.Player = PlayByPlay.Player    
INNER JOIN TeamRating ON Per36Stats.Team = TeamRating.Team 
INNER JOIN PlayerSalary ON Per36Stats.Player = PlayerSalary.Player
INNER JOIN TeamPayroll ON Per36Stats.Team = TeamPayroll.Team

WHERE Per36Stats.Age BETWEEN 21 AND 33
AND PlayerSalary."2027-28" IS NULL
AND Per36Stats.MP > 10
AND Per36Stats.Player NOT IN ('Ben Simmons', 'Russell Westbrook', 'Matisse Thybulle', 'Andre Drummond', 'Tristan Thompson', 'James Harden', 'Chris Paul', 'Kevin Durant', 'Damian Lillard', 'D''Angelo Russell', 'Kyrie Irving')

GROUP BY Per36Stats.Player;




"""


# Execute the query and load results into a DataFrame
Player_Numbers_df = pd.read_sql_query(sql_query, conn)

# Save the merged DataFrame to a new CSV file
csv_filename = "Stats/All_Player_Numbers.csv"
Player_Numbers_df.to_csv(csv_filename, index=False)

# Close the connection
conn.close()

# Provide the file to the user
Player_Numbers_df
```

```python
import pandas as pd
import sqlite3

# Load CSV file into a DataFrame
Player_Numbers_df = pd.read_csv('Stats/All_Player_Numbers.csv')
Team_Rating_df = pd.read_csv('Stats/Team_Ratings.csv')
Team_Payroll_df = pd.read_excel('Stats/Team_Payroll.xlsx')


# Create an SQLite in-memory database
conn = sqlite3.connect(':memory:')

# Load DataFrame into the database as a table
Player_Numbers_df.to_sql('PlayerNumbers', conn, index=False, if_exists='replace')
Team_Rating_df.to_sql('TeamRating', conn, index=False, if_exists='replace')
Team_Payroll_df.to_sql('TeamPayroll', conn, index=False, if_exists='replace')

# Clearly labeling Player Performance Metrics query
player_performance_query = """
SELECT 
    Player, Team,
    "FG%", "3P%", "2P%", "eFG%", "TS%",
    PTS, AST, "AST%",
    ORB, DRB, TRB, "ORB%", "DRB%",
    STL, BLK, "STL%", "BLK%", DBPM,
    TOV, "TOV%",
    "USG%", BPM, "WS/48", VORP, PER
FROM 
PlayerNumbers
"""

# Corrected Team Fit & Trade Value query without duplicates
team_fit_query = """
SELECT 
    Player, TR.Team,
    TR.ORtg, TR.DRtg, TR.NRtg, "On-Off",
    WS, BPM, TR."NRtg/A",
    TP."2024-25_Payroll", TP."2025-26_Payroll", TP."2026-27_Payroll", TP."2027-28_Payroll", TP."2028-29_Payroll", TP."2029-30_Payroll", TP."Luxury_Tax", TP."Penalty",
    "PG%", "SG%", "SF%", "PF%", "C%",
    "3PAr", "FTr"
FROM PlayerNumbers
JOIN TeamRating AS TR ON PlayerNumbers.Team = TR.Team
JOIN TeamPayroll AS TP ON PlayerNumbers.Team = TP.Team
GROUP BY Player
"""

# Corrected Salary & Contract query
salary_contract_query = """
SELECT 
    Player,
    "2024-25", "2025-26", "2026-27", "2027-28", "2028-29", "2029-30", Guaranteed
FROM PlayerNumbers
GROUP BY Player
"""

# Playstyle & Clustering query clearly indicated
playstyle_query = """
SELECT 
    Player,
    BadPass, LostBall,
    OnCourt, "On-Off", DWS, OWS
FROM PlayerNumbers
GROUP BY Player
"""

# Execute the queries and store the results in DataFrames
Player_Performance_df = pd.read_sql_query(player_performance_query, conn)
Team_Rating_Team_Payroll_df = pd.read_sql_query(team_fit_query, conn)
Player_Salary_Contract_df = pd.read_sql_query(salary_contract_query, conn)
Playstyle_Clustering_df = pd.read_sql_query(playstyle_query, conn)

# Close the connection
conn.close()

# Save the extracted data to CSV files
Player_Performance_csv = "Stats/INPUT_Player_Performance.csv"
Team_Rating_Team_Payroll_csv = "Stats/INPUT_Team_Rating_Team_Payroll.csv"
Player_Salary_Contract_csv = "Stats/INPUT_Player_Salary_Contract.csv"
Playstyle_Clustering_csv = "Stats/INPUT_Playstyle_Clustering_Value.csv"

Player_Performance_df.to_csv(Player_Performance_csv, index=False)
Team_Rating_Team_Payroll_df.to_csv(Team_Rating_Team_Payroll_csv, index=False)
Player_Salary_Contract_df.to_csv(Player_Salary_Contract_csv, index=False)
Playstyle_Clustering_df.to_csv(Playstyle_Clustering_csv, index=False)

print("Player_Performace:", Player_Performance_df.head())
print("Team_Rating_Team_Payroll:", Team_Rating_Team_Payroll_df.head())
print("Player_Salary_Contract:", Player_Salary_Contract_df.head())
print("Playstyle_Clustering:", Playstyle_Clustering_df.head())
```

### 1. Identify distinct NBA player roles.



```python
import os
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Explicitly load dataset
Player_Numbers_df = pd.read_csv('Stats/All_Player_Numbers.csv')

# Identify Non-Numeric Columns
non_numeric_cols = Player_Numbers_df.select_dtypes(exclude=["float", "int"]).columns

# Retain 'Player' and 'Team' but remove other non-numeric columns
df_numeric = Player_Numbers_df.drop(
    columns=[col for col in non_numeric_cols if col not in ["Player", "Team"]],
    errors="ignore"
)

# Drop fully NaN columns
df_numeric_filtered = df_numeric.dropna(axis=1, how="all")

# Identify numeric columns for imputation
numeric_columns = df_numeric_filtered.drop(columns=["Player", "Team"], errors="ignore").columns

# Handle missing values
imputer = SimpleImputer(strategy="mean")
X_imputed = imputer.fit_transform(df_numeric_filtered[numeric_columns])

# Confirm column names explicitly
df_numeric_imputed = pd.DataFrame(X_imputed, columns=numeric_columns)

# Explicit Feature Definition (keeping order)
features = [
    # Scoring & Efficiency
    "3PAr", "2P%", "eFG%", "TS%", "FTr", "OWS", "OBPM", "USG%",
    
    # Playmaking & Ball Security
    "AST%", "TOV%", "On-Off", 
    
    # Defense & Rebounding
    "STL%", "BLK%", "DBPM", "TRB%", "DWS",
    
    # Advanced Impact Metrics
    "VORP", "WS/48",
    
    # Position & Fit
    "PG%", "SG%", "SF%", "PF%", "C%"

]

# Ensure all features exist
features = [f for f in features if f in df_numeric_imputed.columns]
if not features:
    raise ValueError("No valid numeric features found for clustering!")

# Extract Player Names and Teams
players = Player_Numbers_df["Player"].reset_index(drop=True) if "Player" in Player_Numbers_df.columns else None
teams = Player_Numbers_df["Team"].reset_index(drop=True) if "Team" in Player_Numbers_df.columns else None

# Data for clustering
X = df_numeric_imputed[features]

# Standardize Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create processed DataFrame, explicitly reset index
processed_df = pd.DataFrame(X_scaled, columns=features).reset_index(drop=True)

# Add Player Names and Teams Back
if players is not None:
    processed_df.insert(0, "Player", players)
if teams is not None:
    processed_df.insert(1, "Team", teams)

# Ensure 'Stats' directory exists
os.makedirs("Stats", exist_ok=True)

# Save to CSV
processed_df.to_csv("Stats/processed_player_data.csv", index=False)

# Reload to confirm
Processed_Players_Df = pd.read_csv("Stats/processed_player_data.csv")

Processed_Players_Df

```

### 2. Selecting the Optimal Number of Clusters

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Load processed dataset
processed_df = pd.read_csv("Stats/processed_player_data.csv")

# Extract features only for clustering
X = processed_df.drop(columns=["Player", "Team"], errors="ignore")

# Range of clusters to test
k_range = range(2, 11)

# Initialize lists to store metrics
inertia = []
silhouette_scores = []
ch_scores = []
db_scores = []

# Calculate metrics for each number of clusters
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X, labels))
    ch_scores.append(calinski_harabasz_score(X, labels))
    db_scores.append(davies_bouldin_score(X, labels))

# Plotting the results
fig, axes = plt.subplots(1, 4, figsize=(24, 5))

# Elbow Method
axes[0].plot(k_range, inertia, marker='o')
axes[0].set_title('Elbow Method')
axes[0].set_xlabel('Number of Clusters')
axes[0].set_ylabel('Inertia')

# Silhouette Scores
axes[1].plot(k_range, silhouette_scores, marker='o')
axes[1].set_title('Silhouette Score')
axes[1].set_xlabel('Number of Clusters')
axes[1].set_ylabel('Silhouette Score')

# Calinski-Harabasz Index
axes[2].plot(k_range, ch_scores, marker='o')
axes[2].set_title('Calinski-Harabasz Index')
axes[2].set_xlabel('Number of Clusters')
axes[2].set_ylabel('CH Score')

# Davies-Bouldin Index
axes[3].plot(k_range, db_scores, marker='o')
axes[3].set_title('Davies-Bouldin Index')
axes[3].set_xlabel('Number of Clusters')
axes[3].set_ylabel('DB Score (lower is better)')

plt.tight_layout()
plt.show()
```

Results

| Metric                    | Best Cluster Choice | Reasoning |
|---------------------------|--------------------|-----------|
| **Elbow Method (Inertia)** | 4 or 5 clusters   | The "elbow" point appears around 4-5, where inertia decreases less sharply. |
| **Silhouette Score**       | 2, then 3-4 clusters | The highest score is at 2, but 3-4 still have relatively high values. |
| **Calinski-Harabasz Index** | 2, then 4 clusters | The highest score is at 2, but 4 maintains a strong score before a steep decline. |
| **Davies-Bouldin Index**   | 6 clusters        | The lowest DB index (better clustering) is at 6, but 4-5 are also reasonable. |
| **Final Recommendation**  | **4 or 5 clusters**    | Balances all metrics well, making it the best overall choice. |

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)
centroids = kmeans.cluster_centers_

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
pca_centroids = pca.transform(centroids)

# Perform t-SNE
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)
tsne_centroids = np.array([X_tsne[clusters == i].mean(axis=0) for i in range(3)])

# Archetype naming for 3 clusters
archetype_map = {
    0: "Role Players & 3/D Wings",
    1: "Creators & Offensive Engines",
    2: "Rim Protectors & Paint Anchors",
}

# Plotting
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
colors = plt.cm.viridis(np.linspace(0.2, 0.8, 3))  # 3 distinct colors

# PCA Plot
for cluster_id in range(3):
    mask = clusters == cluster_id
    axes[0].scatter(X_pca[mask, 0], X_pca[mask, 1],
                    label=archetype_map[cluster_id],
                    alpha=0.7, color=colors[cluster_id])
axes[0].scatter(pca_centroids[:, 0], pca_centroids[:, 1],
                marker='X', c='red', s=200, label='Cluster Centers')
axes[0].set_title('K-Mean Clustering of Player Roles (PCA Visualization)')
axes[0].set_xlabel('Offensive Role (PCA Component 1)')
axes[0].set_ylabel('Defensive Role (PCA Component 2)')
axes[0].legend(loc='best', fontsize='x-small', markerscale=0.6)

# t-SNE Plot
for cluster_id in range(3):
    mask = clusters == cluster_id
    axes[1].scatter(X_tsne[mask, 0], X_tsne[mask, 1],
                    label=archetype_map[cluster_id],
                    alpha=0.7, color=colors[cluster_id])
axes[1].scatter(tsne_centroids[:, 0], tsne_centroids[:, 1],
                marker='X', c='red', s=200, label='Cluster Centers')
axes[1].set_title('K-Mean Clustering of Player Roles (t-SNE Visualization)')
axes[1].set_xlabel('Player Role Axis X (t-SNE)')
axes[1].set_ylabel('Player Role Axis Y (t-SNE)')
axes[1].legend(loc='lower left', fontsize='x-small', markerscale=0.6)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# File paths
player_data_path = "Stats/processed_player_data.csv"
salary_contract_path = "Stats/INPUT_Player_Salary_Contract.csv"
team_rating_payroll_path = "Stats/INPUT_Team_Rating_Team_Payroll.csv"

# Load data
player_data = pd.read_csv(player_data_path)
salary_contract_data = pd.read_csv(salary_contract_path)
team_rating_payroll_data = pd.read_csv(team_rating_payroll_path)

# Simulated values
player_data['MP'] = 1500
player_data['Age'] = 25

# Prepare average team NRtg
team_net_rating_dict = team_rating_payroll_data.groupby("Team")["NRtg"].mean().to_dict()

# Select relevant columns
player_df = player_data[['Player', 'Team', 'OBPM', 'DBPM', 'USG%', 'TS%', 'WS/48', 
                         'TRB%', 'AST%', 'TOV%', 'VORP', 'MP', 'Age']]
salary_df = salary_contract_data[['Player', '2024-25']].copy()
team_df = team_rating_payroll_data[['Player', 'WS', 'BPM', 'NRtg', 'On-Off', 'Luxury_Tax', '2024-25_Payroll']]

# Convert salary to numeric
def salary_to_numeric(salary_str):
    try:
        return float(str(salary_str).replace('$','').replace(',',''))
    except:
        return 0

salary_df['2024-25'] = salary_df['2024-25'].apply(salary_to_numeric)

# Merge datasets
merged_df = player_df.merge(salary_df, on='Player', how='inner').merge(team_df, on='Player', how='inner')

# Value efficiency & new features
merged_df['Value_per_dollar'] = merged_df['WS/48'] / (merged_df['2024-25_Payroll'] + 1e-6)
merged_df["Assist_to_TO"] = merged_df["AST%"] / (merged_df["TOV%"] + 1e-6)
merged_df["Impact_Score"] = merged_df["OBPM"] + merged_df["DBPM"]
merged_df["Efficiency_Value"] = merged_df["TS%"] * merged_df["USG%"]

# 🧠 Scoring function
def calculate_scores(row):
    salary = row['2024-25']
    payroll = row['2024-25_Payroll']

    # ✅ Player Fit Index (original formula)
    fit_index = ((1.5 * row['DBPM']) + (1.5 * row['OBPM']) +
                 (row['USG%'] * 0.8) + (row['AST%'] * 1.2) +
                 (row["TRB%"] * 1.2) + (row['TS%'] * 1.2) -
                 (row['TOV%'] * 1.3)) * 10

    if row['Age']:
        fit_index += max(0, (30 - row['Age'])) * 0.3

    # ✅ Trade Value Index (revised)
    trade_value_index = (row['WS/48'] + row['BPM'] + (0.8 * row['VORP'])) / np.log1p(salary + 1e-6)
    if row['Age']:
        age_factor = max(0, (28 - row['Age'])) * 0.5
        trade_value_index *= (1 + age_factor / 100)
    if row['MP'] < 1000:
        trade_value_index *= 0.9

    # ✅ Net Rating Impact
    team_avg_net_rating = team_net_rating_dict.get(row['Team'], 0)
    relative_net_rating = row['NRtg'] - team_avg_net_rating
    net_rating_impact = ((relative_net_rating * 0.7) + (row['On-Off'] * 1.3)) * row['MP']

    # ✅ Salary Cap Impact (revised)
    cap_impact = (salary / (payroll + 1e-6)) * 100

    return pd.Series({
        "Player Fit Index": fit_index,
        "Trade Value Index": trade_value_index,
        "Net Rating Impact": net_rating_impact,
        "Salary Cap Impact": cap_impact
    })

# Apply score calculation
player_scores = merged_df.apply(calculate_scores, axis=1)

# Final combined DataFrame
final_results = pd.concat([merged_df[['Player', 'Team']], player_scores], axis=1)

# Preview
final_results

```

### Doing R Score 

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

# Start from full dataset
features = merged_df.drop(columns=['Player', 'Team', '2024-25']).copy()

# Feature engineering
features["Assist_to_TO"] = merged_df["AST%"] / (merged_df["TOV%"] + 1e-6)
features["Impact_Score"] = merged_df["OBPM"] + merged_df["DBPM"]
features["Efficiency_Value"] = merged_df["TS%"] * merged_df["USG%"]

# Add salary + financials back in
features["Salary"] = merged_df["2024-25"]
features["Payroll"] = merged_df["2024-25_Payroll"]
features["Luxury_Tax"] = merged_df["Luxury_Tax"]

# Add targets
features["Player Fit Index"] = final_results["Player Fit Index"]
features["Trade Value Index"] = final_results["Trade Value Index"]
features["Salary Cap Impact"] = final_results["Salary Cap Impact"]

# Model all 3 with updated feature set
r2_scores = {}

for target in ["Player Fit Index", "Trade Value Index", "Salary Cap Impact"]:
    data = features.dropna(subset=[target])
    X = data.drop(columns=["Player Fit Index", "Trade Value Index", "Salary Cap Impact"])
    y = data[target]
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

    model = RandomForestRegressor(random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    r2_scores[target] = r2_score(y_test, y_pred)

r2_scores


```

### Cross Validation

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import r2_score

for target in ["Player Fit Index", "Trade Value Index", "Salary Cap Impact"]:
    data = features.dropna(subset=[target])
    X = data.drop(columns=["Player Fit Index", "Trade Value Index", "Salary Cap Impact"])
    y = data[target]

    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
    model = RandomForestRegressor(random_state=42)
    model.fit(X_train, y_train)

    train_r2 = r2_score(y_train, model.predict(X_train))
    test_r2 = r2_score(y_test, model.predict(X_test))
    cv_scores = cross_val_score(model, X, y, cv=10, scoring='r2')

    print(f"\nTarget: {target}")
    print(f"Train R2: {train_r2:.3f}")
    print(f"Test R2:  {test_r2:.3f}")
    print(f"CV mean R2: {cv_scores.mean():.3f}")
```

### Hyperparameter Tuning for All Targets

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
import numpy as np

# Hyperparameter search space
param_dist = {
    'n_estimators': [100, 200, 300, 300],
    'max_depth': [None, 10, 20, 30, 40],
    'min_samples_split': [2, 5, 5],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]
}

# To store results
tuned_results = {}

# Loop through each target
for target in ["Player Fit Index", "Trade Value Index", "Salary Cap Impact"]:
    print(f"\n🔍 Tuning model for: {target}")
    data = features.dropna(subset=[target])
    X = data.drop(columns=["Player Fit Index", "Trade Value Index", "Salary Cap Impact"])
    y = data[target]
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

    rf = RandomForestRegressor(random_state=42)
    random_search = RandomizedSearchCV(
        rf, param_distributions=param_dist,
        n_iter=30, cv=5, scoring='r2',
        n_jobs=-1, random_state=42, verbose=0
    )
    random_search.fit(X_train, y_train)
    best_model = random_search.best_estimator_

    test_r2 = r2_score(y_test, best_model.predict(X_test))
    cv_score = cross_val_score(best_model, X, y, cv=10, scoring='r2').mean()

    tuned_results[target] = {
        "Best Params": random_search.best_params_,
        "Test R2": round(test_r2, 4),
        "CV R2": round(cv_score, 4)
    }

# Display results
import pandas as pd
tuned_df = pd.DataFrame(tuned_results).T
print("\n📊 Tuned Model Results:")
print(tuned_df)
```
