### Using BeautifulSoup to get NBA Advanced & Average Player stats

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL for the 2025 NBA advanced stats
url = "https://www.basketball-reference.com/leagues/NBA_2025_advanced.html"

# Set headers to mimic a real browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}

# Get the webpage content
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Locate the correct table (ID: "advanced")
table = soup.find("table", {"id": "advanced"})

# Check if the table exists
if table:
    # Extract headers
    headers = [th.get_text() for th in table.find("thead").find_all("th")]

    # Extract rows
    rows = []
    for row in table.find("tbody").find_all("tr"):
        cols = [td.get_text() for td in row.find_all("td")]
        if cols:  # Ignore empty rows
            rows.append(cols)

    # Convert to DataFrame
    df = pd.DataFrame(rows, columns=headers[1:])  # Skip the rank column

    # Show the top rows
    print(df.head())

    # Save the data to a CSV file
    df.to_csv("NBA_Advanced_Stats_2025.csv", index=False)
    print("Data saved to nba_advanced_stats_2025.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL for NBA Team Ratings
url = "https://www.basketball-reference.com/leagues/NBA_2025_ratings.html"

# Set headers to mimic a real browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}

# Get the webpage content
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Locate the correct table
table = soup.find("table", {"id": "ratings"})

# Debugging: Print the table HTML if found
if table:
    print("Table found! Extracting data...\n")
    print(table.prettify()[:1000])  # Print the first 1000 characters of the table for inspection

    # Extract headers
    header_th = table.find("thead").find_all("th")
    headers = [th.get_text(strip=True) for th in header_th]
    print(f"Extracted headers: {headers}")  # Debugging step

    # Extract rows
    rows = []
    for row in table.find("tbody").find_all("tr"):
        cols = [td.get_text(strip=True) for td in row.find_all("td")]

        # Only append rows with data
        if cols:
            rows.append(cols)

    # Debugging: Check if rows are extracted
    if not rows:
        print("⚠️ No rows extracted! The table might be empty, hidden, or JavaScript-rendered.")
    else:
        print(f"✅ Extracted {len(rows)} rows successfully.")  # Debugging step
        print(f"First row sample: {rows[0]}")  # Print first row to verify data

    # Convert to DataFrame only if data is available
    if rows:
        # Ensure headers match row length
        headers = headers[:len(rows[0])]  # Trim to match row column count

        df = pd.DataFrame(rows, columns=headers)  # Match columns dynamically

        # Show first few rows
        print(df.head())

        # Save to CSV
        df.to_csv("NBA_Team_Needs_2025.csv", index=False)
        print("✅ Data saved to NBA_Team_Needs_2025.csv")
else:
    print("⚠️ Table with id='ratings' NOT found. Check if the page structure has changed or data is loaded via JavaScript.")


```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the target URL
url = "https://www.spotrac.com/nba/cap/"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# Send the request
response = requests.get(url, headers=headers)
if response.status_code != 200:
    print(f"Failed to retrieve data: {response.status_code}")
    exit()

# Parse the HTML
soup = BeautifulSoup(response.text, "html.parser")

# Find the table
table = soup.find("table", {"class": "datatable"})
if not table:
    print("Table not found. The website structure may have changed.")
    exit()

# Extract data
data = []
rows = table.find_all("tr")[1:]  # Skip the header row

for row in rows:
    cols = row.find_all("td")
    if len(cols) > 1:
        team = cols[0].get_text(strip=True)
        salary_cap = cols[1].get_text(strip=True)
        data.append({"Team": team, "Salary Cap": salary_cap})

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("NBA_Salary_Cap.csv", index=False)

# Display the DataFrame
print(df)

```
