## Merging Player Data

```python
import pandas as pd
import sqlite3

# Load CSV files into DataFrames
Time_Per_Game_df = pd.read_csv('Stats/Per_36_min_Stats.csv')
Advanced_Stats_df = pd.read_csv('Stats/Advanced_Stats.csv')
Poss_Per_Game_df = pd.read_csv('Stats/Per_100_Poss_Stats.csv')
Play_by_Play_df = pd.read_csv('Stats/Play_by_Play.csv')
Player_Salary_df = pd.read_csv('Stats/Player_Salary.csv')
Team_Rating_df = pd.read_csv('Stats/Team_Ratings.csv')

# Create an SQLite in-memory database
conn = sqlite3.connect(':memory:')

# Load DataFrames into the database as tables
Time_Per_Game_df.to_sql('Per36Stats', conn, index=False, if_exists='replace')
Advanced_Stats_df.to_sql('AdvancedStats', conn, index=False, if_exists='replace')
Poss_Per_Game_df.to_sql('Per100PossStats', conn, index=False, if_exists='replace')
Play_by_Play_df.to_sql('PlayByPlay', conn, index=False, if_exists='replace')
Player_Salary_df.to_sql('PlayerSalary', conn, index=False, if_exists='replace')
Team_Rating_df.to_sql('TeamRating', conn, index=False, if_exists='replace')


# SQL query to merge all tables on 'Player'
sql_query = """
    SELECT Per36Stats.*, AdvancedStats.*, Per100PossStats.*, PlayByPlay.*, PlayerSalary.*, TeamRating.*
    FROM Per36Stats
    INNER JOIN AdvancedStats ON Per36Stats.Player = AdvancedStats.Player
    INNER JOIN Per100PossStats ON Per36Stats.Player = Per100PossStats.Player
    INNER JOIN PlayByPlay ON Per36Stats.Player = PlayByPlay.Player    
    INNER JOIN TeamRating ON Per36Stats.Team = TeamRating.Team 
    INNER JOIN PlayerSalary ON Per36Stats.Player = PlayerSalary.Player
    AND Per36Stats.Age < 35
    AND Per36Stats.Age > 20
    AND PlayerSalary.[2027-28] IS NULL
    AND PlayerSalary.[2028-29] IS NULL
    AND PlayerSalary.[2029-30] IS NULL
    AND PlayerSalary.[2026-27] IS NULL
    GROUP BY Per36Stats.Player
    """

# Execute the query and load results into a DataFrame
Player_Numbers_df = pd.read_sql_query(sql_query, conn)

# Step 1: Fix Duplicate Column Names

# Save the merged DataFrame to a new CSV file
csv_filename = "Stats/All_Player_Numbers.csv"
Player_Numbers_df.to_csv(csv_filename, index=False)

# Close the connection
conn.close()

# Provide the file to the user
Player_Numbers_df

```

## **Inputs: Features Considered for Trade Suggestions**
| **Category**                    | **Metrics** |
|---------------------------------|------------|
| **A. Player Performance Metrics** | **Scoring**: FG%, 3P%, 2P%, eFG%, TS%, PTS <br> **Playmaking**: AST, AST%, TOV, TOV% <br> **Rebounding**: ORB, DRB, TRB, ORB%, DRB% <br> **Defense**: STL, BLK, STL%, BLK%, DBPM <br> **Usage Rate & Impact**: USG%, BPM, WS/48, VORP |
| **B. Team Fit & Trade Value**   | **Offensive/Defensive Metrics**: ORtg, DRtg, NRtg, On-Off <br> **Trade Fairness Indicators**: <br> - Change in BPM, WS before/after trade <br> - Net Rating Impact (`NRtg/A`) <br> - Positional Fit (PG%, SG%, SF%, PF%, C%) |

| **D. Playstyle & Clustering Attributes** | **Turnover Tendencies**: `BadPass`, `LostBall` <br> **Playstyle Variations**: `OnCourt`, `On-Off`, `DWS`, `OWS` |


```python
import pandas as pd
import sqlite3

# Load CSV file into a DataFrame
Player_Numbers_df = pd.read_csv('Stats/All_Player_Numbers.csv')
Team_Rating_df = pd.read_csv('Stats/Team_Ratings.csv')


# Create an SQLite in-memory database
conn = sqlite3.connect(':memory:')

# Load DataFrame into the database as a table
Player_Numbers_df.to_sql('PlayerNumbers', conn, index=False, if_exists='replace')
Team_Rating_df.to_sql('TeamRating', conn, index=False, if_exists='replace')


# SQL query to extract features for Team Fit & Trade Value
player_performance_query = """
SELECT 
    Player,
    "FG%", "3P%", "2P%", "eFG%", "TS%",
    PTS, AST, "AST%",
    ORB, DRB, TRB, "ORB%", "DRB%",
    STL, BLK, "STL%", "BLK%", DBPM,
    TOV, "TOV%",
    "USG%", BPM, "WS/48", VORP, PER
FROM 
PlayerNumbers
"""

# SQL query to extract features for Team Fit & Trade Value
team_fit_query = """
SELECT 
    Player,
    TR.ORtg, TR.DRtg, TR.NRtg, "On-Off",
    WS, BPM, TR."NRtg/A",
    "PG%", "SG%", "SF%", "PF%", "C%",
    "3PAr", "FTr"
FROM PlayerNumbers
JOIN TeamRating AS TR ON PlayerNumbers.Team = TR.Team
GROUP BY PlayerNumbers.Player
"""

# SQL query to extract features for Salary & Contract
salary_contract_query = """
SELECT 
    Player,
    "2024-25", "2025-26", Guaranteed
FROM PlayerNumbers
GROUP BY PlayerNumbers.Player
"""

# SQL query to extract features for Playstyle & Clustering
playstyle_query = """
SELECT 
    Player,
    BadPass, LostBall,
    OnCourt, "On-Off", DWS, OWS
FROM PlayerNumbers
GROUP BY PlayerNumbers.Player
"""

# Execute the queries and store the results in DataFrames
Player_Performance_df = pd.read_sql_query(player_performance_query, conn)
Team_Fit_Value_df = pd.read_sql_query(team_fit_query, conn)
Player_Salary_Contract_df = pd.read_sql_query(salary_contract_query, conn)
Playstyle_Clustering_df = pd.read_sql_query(playstyle_query, conn)

# Close the connection
conn.close()

# Save the extracted data to CSV files
Player_Performance_csv = "Stats/INPUT_Player_Performance.csv"
Team_Fit_Value_csv = "Stats/INPUT_Team_Fit_Value.csv"
Player_Salary_Contract_csv = "Stats/INPUT_Player_Salary_Contract.csv"
Playstyle_Clustering_csv = "Stats/INPUT_Playstyle_Clustering_Value.csv"

Player_Performance_df.to_csv(Player_Performance_csv, index=False)
Team_Fit_Value_df.to_csv(Team_Fit_Value_csv, index=False)
Player_Salary_Contract_df.to_csv(Player_Salary_Contract_csv, index=False)
Playstyle_Clustering_df.to_csv(Playstyle_Clustering_csv, index=False)

print(Player_Performance_df.columns)
print(Team_Fit_Value_df.columns)
print(Player_Salary_Contract_df.columns)
print(Playstyle_Clustering_df.columns)
```

## 1. **Determining the Number of Clusters needed**
1. Elbow Method (Inertia)
2. Silhouette Score (SS)
3. Calinski-Harabasz Index (CHI)
4. Davies-Bouldin Index (DBI)
5. Gap Statistic: Compares WCSS to a null reference distribution.


Based on the graphs provided, the most optimal **number of clusters (k)** appears to be **5 or 6**, as determined by these key observations:

### **Choosing the Optimal Number of Clusters (\( k \))**

#### **1. Elbow Method (Inertia)**
- The largest drops in inertia occur from **\( k=2 \) to \( k=6 \)**.
- After \( k=7 \), the inertia decreases at a slower rate.
- The "elbow" appears to be **around \( k=6 \) or \( k=7 \)**.

#### **2. Silhouette Score**
- The highest silhouette score appears at **\( k=4 \) (0.152)**.
- \( k=7 \) and \( k=11 \) also have relatively higher silhouette scores.
- A silhouette score above **0.10–0.15 is considered weak but usable**, meaning clusters are not very well separated.

#### **3. Calinski-Harabasz Index**
- The CH Index starts high at \( k=2 \), but steadily declines.
- The value stabilizes around **\( k=7 \) to \( k=8 \)**.
- Higher values indicate better-defined clusters, and here, **\( k=7 \) has a slight local peak**.

#### **4. Davies-Bouldin Index (DBI)**
- Lower values are better.
- The best DBI values are at **\( k=12 \) (1.38) and \( k=13 \) (1.27)**.
- **\( k=8 \) and \( k=7 \) also have lower values**, suggesting that clusters are reasonably well-separated.

---

### **Final Recommendation:**
1. **If you want a balance between distinctiveness and interpretability** → **\( k=6 \) or \( k=7 \)**.
2. **If you want more detailed role-based clusters** → **\( k=8 \) to \( k=10 \)**.
3. **If you want maximal cluster separation (better but harder to interpret)** → **\( k=12 \) to \( k=13 \)**.

💡 **Best choice:** **\( k=7 \)** aligns with distinct roles like **Scoring Guards, 3&D Wings, Playmakers, Defensive Bigs, Scoring Bigs, Role Players, and All-Stars/Superstars.**

Best K=7


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

# Load datasets
Player_Numbers_df

# Define relevant features based on cluster characteristics
features = [
    "PTS", "TS%", "FG%", "3P%", "2P%", "3PAr", "FT%", "eFG%", "FTr", "USG%", "ORtg", "DRtg",  # Scoring & Efficiency
    "AST", "AST%", "TOV", "TOV%", "BadPass", "LostBall",  # Playmaking & Turnovers
    "ORB%", "DRB%", "PER",  # Rebounding
    "STL", "STL%", "BLK", "BLK%", "DBPM", "DWS", "OWS",  # Defense
    "BPM", "WS/48", "VORP", "OnCourt", "On-Off", "NRtg/A",  # Advanced Impact Metrics
    "PG%", "SG%", "SF%", "PF%", "C%"  # Positional Play
     # Determine if Role Player
]


# Extract player names
players = Player_Numbers_df["Player"]
X = Player_Numbers_df[features]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Determine the optimal number of clusters using the Elbow Method
from sklearn.impute import SimpleImputer

# Create an imputer that replaces NaNs with the mean of the respective column
imputer = SimpleImputer(strategy="mean")  
X_scaled = imputer.fit_transform(X_scaled)

inertia = []
k_range = range(1, 15)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(6, 3))
plt.plot(k_range, inertia, marker="o")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.show()

ebm_df = pd.DataFrame({"Number of Clusters (k)": list(k_range), "Elbow Method": inertia})
print(ebm_df)
```

```python
from sklearn.metrics import silhouette_score

# Calculate silhouette scores for different cluster sizes
silhouette_scores = []
k_range = range(2, 15)  

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_scaled)
    silhouette_avg = silhouette_score(X_scaled, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot the silhouette scores
plt.figure(figsize=(6, 3))
plt.plot(k_range, silhouette_scores, marker="o", linestyle="-")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for Optimal k")
plt.show()

sss_df = pd.DataFrame({"Number of Clusters (k)": list(k_range), "Silhouette Score": silhouette_scores})
print(sss_df)
```

```python
import numpy as np
from sklearn.metrics import calinski_harabasz_score

# Assume X_scaled is the preprocessed dataset

# Calculate Calinski-Harabasz Index for different cluster sizes
chi_scores = []
k_range = range(2, 15)  

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_scaled)
    chi_score = calinski_harabasz_score(X_scaled, cluster_labels)
    chi_scores.append(chi_score)

# Plot the CHI scores
plt.figure(figsize=(6, 3))
plt.plot(k_range, chi_scores, marker="o", linestyle="-")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Calinski-Harabasz Index")
plt.title("Calinski-Harabasz Index for Optimal k")
plt.show()

chi_df = pd.DataFrame({"Number of Clusters (k)": list(k_range), "Calinski-Harabasz Index": chi_scores})
print(chi_df)
```

```python
# Calculate Davies-Bouldin Index for different cluster sizes

from sklearn.metrics import davies_bouldin_score

dbi_scores = []
k_range = range(2, 15)  

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_scaled)
    dbi_score = davies_bouldin_score(X_scaled, cluster_labels)
    dbi_scores.append(dbi_score)

# Plot the DBI scores
plt.figure(figsize=(6, 3))
plt.plot(k_range, dbi_scores, marker="o", linestyle="-")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Davies-Bouldin Index")
plt.title("Davies-Bouldin Index for Optimal k")
plt.show()

dbi_df = pd.DataFrame({"Number of Clusters (k)": list(k_range), "Davies-Bouldin Index": dbi_scores})
print(dbi_df)
```

```python
# Apply KMeans clustering
kmeans = KMeans(n_clusters=7, random_state=42, n_init=10)
Player_Numbers_df["Cluster"] = kmeans.fit_predict(X_scaled)

# Convert cluster centers back to the original scale
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)

# Create a DataFrame for easy interpretation
cluster_df = pd.DataFrame(cluster_centers, columns=features)
cluster_df.index = [f"Cluster {i}" for i in range(7)]
print(cluster_df)
```

```python
# Group players by cluster and display
for cluster in range(7):
    print(f"Cluster {cluster} Players:")
    print(Player_Numbers_df[Player_Numbers_df["Cluster"] == cluster][["Player"]].head(10))  # Show top 10 for each
    print("-" * 50)
```

## 2. Apply Clustering Algorithm

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Place the K-Cluster to be 7
optimal_k = 7

# Perform PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Add PCA results to DataFrame
Player_Numbers_df["PCA1"] = X_pca[:, 0]
Player_Numbers_df["PCA2"] = X_pca[:, 1]
```

### A. **PCA Visualization for K-Means Clusters**

```python
# Apply K-Means Clustering
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
Player_Numbers_df["Cluster_KMeans"] = kmeans.fit_predict(X_scaled)

# Plot K-Means Clusters
plt.figure(figsize=(8, 6))
sns.scatterplot(x=Player_Numbers_df["PCA1"], y=Player_Numbers_df["PCA2"], 
                hue=Player_Numbers_df["Cluster_KMeans"], palette="tab10", alpha=0.7)
plt.title("K-Means Clusters Visualized with PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster")
plt.show()
```

### B. **PCA Visualization for Hierarchical Clustering**

```python
# Apply Hierarchical Clustering
hierarchical = AgglomerativeClustering(n_clusters=optimal_k, linkage="ward")
Player_Numbers_df["Cluster_Hierarchical"] = hierarchical.fit_predict(X_scaled)


plt.figure(figsize=(8, 6))
sns.scatterplot(x=Player_Numbers_df["PCA1"], y=Player_Numbers_df["PCA2"], 
                hue=Player_Numbers_df["Cluster_Hierarchical"], palette="tab10", alpha=0.7)
plt.title("Hierarchical Clustering Visualized with PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster")
plt.show()
```

### C. **PCA Visualization for Gaussian Mixture Model**

```python
# Apply Gaussian Mixture Model (GMM)
gmm = GaussianMixture(n_components=optimal_k, random_state=42)
Player_Numbers_df["Cluster_GMM"] = gmm.fit_predict(X_scaled)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=Player_Numbers_df["PCA1"], y=Player_Numbers_df["PCA2"], 
                hue=Player_Numbers_df["Cluster_GMM"], palette="tab10", alpha=0.7)

plt.title("GMM Clusters Visualized with PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.legend(title="Cluster")
plt.show()
```

### D. **TSNE Visualization for All Clusters**

```python
from sklearn.manifold import TSNE

# Perform t-SNE
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_tsne = tsne.fit_transform(X_scaled)

# Add t-SNE results to DataFrame
Player_Numbers_df["tSNE1"] = X_tsne[:, 0]
Player_Numbers_df["tSNE2"] = X_tsne[:, 1]

# Plot t-SNE Clustering
plt.figure(figsize=(8, 6))
sns.scatterplot(x=Player_Numbers_df["tSNE1"], y=Player_Numbers_df["tSNE2"], 
                hue=Player_Numbers_df["Cluster_KMeans"], palette="tab10", alpha=0.7)
plt.title("t-SNE Visualization of K-Means Clusters")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.legend(title="Cluster")
plt.show()
```
