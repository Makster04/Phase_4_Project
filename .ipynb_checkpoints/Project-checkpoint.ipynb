## Merging Player Data

```python
import pandas as pd
import sqlite3

# Load CSV files into DataFrames
Time_Per_Game_df = pd.read_csv('Stats/Per_36_min_Stats.csv')
Advanced_Stats_df = pd.read_csv('Stats/Advanced_Stats.csv')
Poss_Per_Game_df = pd.read_csv('Stats/Per_100_Poss_Stats.csv')
Play_by_Play_df = pd.read_csv('Stats/Play_by_Play.csv')
Player_Salary_df = pd.read_csv('Stats/Player_Salary.csv')
Team_Rating_df = pd.read_csv('Stats/Team_Ratings.csv')
Team_Payroll_df = pd.read_excel('Stats/Team_Payroll.xlsx')


# Create an SQLite in-memory database
conn = sqlite3.connect(':memory:')

# Load DataFrames into the database as tables
Time_Per_Game_df.to_sql('Per36Stats', conn, index=False, if_exists='replace')
Advanced_Stats_df.to_sql('AdvancedStats', conn, index=False, if_exists='replace')
Poss_Per_Game_df.to_sql('Per100PossStats', conn, index=False, if_exists='replace')
Play_by_Play_df.to_sql('PlayByPlay', conn, index=False, if_exists='replace')
Player_Salary_df.to_sql('PlayerSalary', conn, index=False, if_exists='replace')
Team_Rating_df.to_sql('TeamRating', conn, index=False, if_exists='replace')
Team_Payroll_df.to_sql('TeamPayroll', conn, index=False, if_exists='replace')



# SQL query to merge all tables on 'Player'
sql_query = """
    SELECT Per36Stats.*, AdvancedStats.*, Per100PossStats.*, PlayByPlay.*, 
           PlayerSalary.*, TeamRating.*, TeamPayroll.*
    FROM Per36Stats
    INNER JOIN AdvancedStats ON Per36Stats.Player = AdvancedStats.Player
    INNER JOIN Per100PossStats ON Per36Stats.Player = Per100PossStats.Player
    INNER JOIN PlayByPlay ON Per36Stats.Player = PlayByPlay.Player    
    INNER JOIN TeamRating ON Per36Stats.Team = TeamRating.Team 
    INNER JOIN PlayerSalary ON Per36Stats.Player = PlayerSalary.Player
    INNER JOIN TeamPayroll ON Per36Stats.Team = TeamPayroll.Team
    WHERE Per36Stats.Age BETWEEN 21 AND 34
    AND PlayerSalary.[2027-28] IS NULL
    AND PlayerSalary.[2028-29] IS NULL
    AND PlayerSalary.[2029-30] IS NULL
    AND PlayerSalary.[2026-27] IS NULL
    AND Per36Stats.Player NOT IN ('Ben Simmons', 'Russell Westbrook', 'James Harden', 'Chris Paul', 'Kevin Durant', 'Damian Lillard')
    
    -- Trade Value & Performance Criteria
    AND (
        (AdvancedStats.[BPM] > 2 OR AdvancedStats.[BPM] < -1)
        OR (AdvancedStats.[VORP] > 1 OR AdvancedStats.[VORP] < 0)
        OR (AdvancedStats.[USG%] > 20)
        OR (AdvancedStats.[WS] > 3 OR AdvancedStats.[WS] < 1)
        OR (TeamRating.[NRtg] < -3 OR TeamRating.[NRtg] > 5)
        OR (PlayByPlay.[OnCourt] < 1000 OR (PlayByPlay.[On-Off] >= -2 AND PlayByPlay.[On-Off] <= 5))
    )
    -- Financial Considerations Based on 2024-25 Payroll
    AND (
        -- Expiring or overpaid contracts
        (PlayerSalary.[2025-26] IS NULL OR PlayerSalary.[2024-25] > 20000000 AND AdvancedStats.[BPM] < 0)
        -- High-spending teams ($180M+) likely to shed salary
        OR (TeamPayroll.[2024-25_Payroll] > 180000000)
        -- Low-spending teams (< $160M) likely to acquire contracts
        OR (TeamPayroll.[2024-25_Payroll] < 160000000)
    )
    GROUP BY Per36Stats.Player;


"""


# Execute the query and load results into a DataFrame
Player_Numbers_df = pd.read_sql_query(sql_query, conn)


# Save the merged DataFrame to a new CSV file
csv_filename = "Stats/All_Player_Numbers.csv"
Player_Numbers_df.to_csv(csv_filename, index=False)

# Close the connection
conn.close()

# Provide the file to the user
Player_Numbers_df
```

```python
import pandas as pd
import sqlite3

# Load CSV file into a DataFrame
Player_Numbers_df = pd.read_csv('Stats/All_Player_Numbers.csv')
Team_Rating_df = pd.read_csv('Stats/Team_Ratings.csv')
Team_Payroll_df = pd.read_excel('Stats/Team_Payroll.xlsx')


# Create an SQLite in-memory database
conn = sqlite3.connect(':memory:')

# Load DataFrame into the database as a table
Player_Numbers_df.to_sql('PlayerNumbers', conn, index=False, if_exists='replace')
Team_Rating_df.to_sql('TeamRating', conn, index=False, if_exists='replace')
Team_Payroll_df.to_sql('TeamPayroll', conn, index=False, if_exists='replace')

# SQL query to extract features for Team Fit & Trade Value
player_performance_query = """
SELECT 
    Player, Team,
    "FG%", "3P%", "2P%", "eFG%", "TS%",
    PTS, AST, "AST%",
    ORB, DRB, TRB, "ORB%", "DRB%",
    STL, BLK, "STL%", "BLK%", DBPM,
    TOV, "TOV%",
    "USG%", BPM, "WS/48", VORP, PER
FROM 
PlayerNumbers
"""

# SQL query to extract features for Team Fit & Trade Value
team_fit_query = """
SELECT 
    Player, 
    TR.ORtg, TR.DRtg, TR.NRtg, "On-Off",
    WS, BPM, TR."NRtg/A", TP."2024-25_Payroll", TP."2025-26_Payroll", TP."2024-25_Payroll", TP."2025-26_Payroll", TP."2026-27_Payroll", TP."2027-28_Payroll", TP."2028-29_Payroll", TP."2029-30_Payroll", TP."Luxury_Tax", TP."Penalty",
    "PG%", "SG%", "SF%", "PF%", "C%",
    "3PAr", "FTr"
FROM PlayerNumbers
JOIN TeamRating AS TR ON PlayerNumbers.Team = TR.Team
JOIN TeamPayroll AS TP ON PlayerNumbers.Team = TP.Team
GROUP BY PlayerNumbers.Player
"""

# SQL query to extract features for Salary & Contract
salary_contract_query = """
SELECT 
    Player,
    "2024-25", "2025-26", "2024-25", "2025-26", "2026-27", "2027-28", "2028-29", "2029-30",  Guaranteed
FROM PlayerNumbers
GROUP BY PlayerNumbers.Player
"""

# SQL query to extract features for Playstyle & Clustering
playstyle_query = """
SELECT 
    Player,
    BadPass, LostBall,
    OnCourt, "On-Off", DWS, OWS
FROM PlayerNumbers
GROUP BY PlayerNumbers.Player
"""

# Execute the queries and store the results in DataFrames
Player_Performance_df = pd.read_sql_query(player_performance_query, conn)
Team_Rating_Team_Payroll_df = pd.read_sql_query(team_fit_query, conn)
Player_Salary_Contract_df = pd.read_sql_query(salary_contract_query, conn)
Playstyle_Clustering_df = pd.read_sql_query(playstyle_query, conn)

# Close the connection
conn.close()

# Save the extracted data to CSV files
Player_Performance_csv = "Stats/INPUT_Player_Performance.csv"
Team_Rating_Team_Payroll_csv = "Stats/INPUT_Team_Rating_Team_Payroll.csv"
Player_Salary_Contract_csv = "Stats/INPUT_Player_Salary_Contract.csv"
Playstyle_Clustering_csv = "Stats/INPUT_Playstyle_Clustering_Value.csv"

Player_Performance_df.to_csv(Player_Performance_csv, index=False)
Team_Rating_Team_Payroll_df.to_csv(Team_Rating_Team_Payroll_csv, index=False)
Player_Salary_Contract_df.to_csv(Player_Salary_Contract_csv, index=False)
Playstyle_Clustering_df.to_csv(Playstyle_Clustering_csv, index=False)

print("Player_Performace:", Player_Performance_df.columns)
print("Team_Rating_Team_Payroll:", Team_Rating_Team_Payroll_df.columns)
print("Player_Salary_Contract:", Player_Salary_Contract_df.columns)
print("Playstyle_Clustering:", Playstyle_Clustering_df.columns)
```

## **Best Number of Clusters Based on the New Metrics**

Here’s a **formatted table** summarizing the clustering interpretation and recommendations:

| Method | Goal | Observations | Best k |
|--------|------|-------------|--------|
| 1️⃣ **Elbow Method** | Find the "elbow" where inertia stops decreasing significantly. | Large drops in inertia up to **k = 5 (1460.62)**, then a more gradual decline. After **k = 5**, the decrease slows down. | **5** |
| 2️⃣ **Silhouette Score** | Measures how well-separated clusters are (**higher is better**). | Best score at **k = 2 (0.178)**. Also high at **k = 7 (0.150)** and **k = 10 (0.151)**. **k = 5 (0.126)** is lower but acceptable. | **2, 7, or 10 (k = 5 is still viable)** |
| 3️⃣ **Calinski-Harabasz Index (CH Index)** | Measures how well-separated clusters are (**higher is better**). | Highest value at **k = 2 (22.86)**. **k = 3–5** have reasonable values (**20.65 to 16.13**). After **k = 5**, the values decline. | **3, 4, or 5** |
| 4️⃣ **Davies-Bouldin Index (DB Index)** | Measures **cluster compactness** (**lower is better**). | Best value at **k = 10 (1.38)**. **k = 7 (1.49)** and **k = 5 (1.81)** are also reasonable. Higher values (**10-13**) perform better but may over-cluster. | **7 or 10 (k = 5 is reasonable)** |

#### **✅ Final Recommendation: k = 5**

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer

# Load dataset (ensure this is done before using it)
Player_Numbers_df 

# Step 1: Identify Non-Numeric Columns
non_numeric_cols = Player_Numbers_df.select_dtypes(exclude=["float", "int"]).columns

# Step 2: Retain 'Player' and 'Team' but remove other non-numeric columns
df_numeric = Player_Numbers_df.drop(columns=[col for col in non_numeric_cols if col not in ["Player", "Team"]], errors="ignore")

# Step 3: Drop Columns That Are Fully NaN Before Imputation
df_numeric_filtered = df_numeric.dropna(axis=1, how="all")  # Drop columns that are entirely NaN

# Step 4: Identify the numeric columns for imputation
numeric_columns = df_numeric_filtered.drop(columns=["Player", "Team"], errors="ignore").columns

# Step 5: Handle Missing Values
imputer = SimpleImputer(strategy="mean")  
X_imputed = imputer.fit_transform(df_numeric_filtered[numeric_columns])

# Step 6: Ensure Column Names Match After Imputation
df_numeric_imputed = pd.DataFrame(X_imputed, columns=numeric_columns[:X_imputed.shape[1]])

# Step 7: Define Features for Clustering (Removing Duplicates)
features = list(set([
    # Scoring & Efficiency
    "3P%", "2P%", "eFG%", "TS%", "USG%", "3PAr", "FTr", "OWS", "OBPM",
    
    # Playmaking & Ball Security
    "AST%", "TOV%", "BadPass", "LostBall", "OnCourt", "On-Off",
    
    # Defense & Rebounding
    "STL%", "BLK%", "DBPM", "DRB%", "ORB%", "DWS",
    
    # Advanced Impact Metrics
    "VORP", "PER",
    
    # Position & Fit
    "PG%", "SG%", "SF%", "PF%", "C%"
]))

# Ensure all features exist in the dataset
features = [f for f in features if f in df_numeric_imputed.columns]
if not features:
    raise ValueError("No valid numeric features found for clustering!")

# Extract Player Names, Teams & Feature Data
players = Player_Numbers_df["Player"] if "Player" in Player_Numbers_df.columns else None
teams = Player_Numbers_df["Team"] if "Team" in Player_Numbers_df.columns else None

X = df_numeric_imputed[features]

# Step 8: Standardize Features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 9: Create a new DataFrame with the processed data
processed_df = pd.DataFrame(X_scaled, columns=features)

# Step 10: Add Player Names and Teams Back
if players is not None:
    processed_df.insert(0, "Player", players.values)
if teams is not None:
    processed_df.insert(1, "Team", teams.values)

# Step 11: Ensure 'Stats' directory exists
os.makedirs("Stats", exist_ok=True)

# Save to CSV
Scaled_Players = "Stats/processed_player_data.csv"
processed_df.to_csv(Scaled_Players, index=False)

# Reload processed data
Processed_Players_Df = pd.read_csv(Scaled_Players)

Processed_Players_Df
```

## 2. Apply Clustering Algorithm

```python
# Tools
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
from sklearn.manifold import TSNE
```

## 1. **SHAP (SHapley Addictive Explinations)**
SHAP helps explain which features contribute most to trade suggestions. You can implement it like this:

```python
import pandas as pd
import numpy as np
import shap
import matplotlib.pyplot as plt
import networkx as nx
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import cosine_similarity

# Load the dataset
file_path = "Stats/Cleaned_Trade_Evaluation_Data.csv"
trade_data = pd.read_csv(file_path)

# Selecting relevant features from the dataset
features = ['eFG%', 'TS%', 'USG%', 'AST%', 'TOV%', 'STL%', 'BLK%', 'ORtg', 'DRtg', 'NRtg', 'Value_per_Dollar']
target_variable = 'Cap_Impact_Score'

# Dropping rows with missing values in selected features
trade_data = trade_data.dropna(subset=features + [target_variable])

# Standardizing the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(trade_data[features])

# Training a RandomForestRegressor for SHAP analysis
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(scaled_features, trade_data[target_variable])

# Compute SHAP values
explainer = shap.TreeExplainer(model)
shap_values = explainer.shap_values(scaled_features)

# Display SHAP summary plot
shap.summary_plot(shap_values, trade_data[features])

# Trade Metrics Summary
trade_metrics = trade_data[['Player']].copy()
trade_metrics['Player Fit Score'] = trade_data['Value_per_Dollar']  # Placeholder metric
trade_metrics['Trade Value Score'] = trade_data['NRtg']  # Using Net Rating as trade fairness proxy
trade_metrics['Net Rating Impact'] = trade_data['NRtg'] / trade_data['USG%']  # Fixing Net Rating per Attempt
trade_metrics['Salary Cap Impact Score'] = trade_data['Cap_Impact_Score']
trade_metrics['On-Off Impact'] = (trade_data['ORtg'] - trade_data['DRtg']) / trade_data['USG%']

# Display trade metrics
trade_metrics
```

```python
from sklearn.inspection import permutation_importance

# Compute permutation importance
perm_importance = permutation_importance(model, scaled_features, trade_data[target_variable], n_repeats=10, random_state=42)

# Extract and visualize feature importance
feature_importance = dict(zip(features, perm_importance.importances_mean))
plt.figure(figsize=(10, 6))
plt.barh(list(feature_importance.keys()), list(feature_importance.values()), color='skyblue')
plt.xlabel("Feature Importance")
plt.ylabel("Trade Evaluation Features")
plt.title("Permutation-Based Feature Importance (Trade Evaluation)")
plt.show()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import networkx as nx

# Creating a graph-based trade network using cosine similarity
G = nx.Graph()

# Compute player similarity based on stats
similarity_matrix = cosine_similarity(scaled_features)
players = trade_data['Player'].values

# Create trade connections based on highest similarity
num_connections = 100  # Number of trade connections to form

for idx_a in range(len(players)):
    # Find the most similar player
    idx_b = np.argsort(similarity_matrix[idx_a])[-2]  # Exclude self

    player_a, player_b = players[idx_a], players[idx_b]
    trade_value = similarity_matrix[idx_a][idx_b]  # Use similarity score as trade value
    G.add_edge(player_a, player_b, weight=trade_value)


# Visualizing the trade impact network
fig, ax = plt.subplots(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=200, font_size=8, edge_color="gray", alpha=0.7, ax=ax)
plt.title("Trade Impact Network Based on Player Similarity")
plt.show()
```

## 2. Determining best number of clusters:

```python
# Re-import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

import os
os.environ["OMP_NUM_THREADS"] = "1"


# Reload the dataset
file_path = "Stats/Cleaned_Trade_Evaluation_Data.csv"
trade_data = pd.read_csv(file_path)

# Selecting relevant features for clustering (Expanded with Player & Team-Level Metrics)
features = [
    # Shooting & Scoring
    'eFG%', 'TS%', 'USG%', '3PAr', 'FTr',

    # Playmaking & Turnovers
    'AST%', 'TOV%', 'BadPass', 'LostBall',

    # Defense & Rebounding
    'STL%', 'BLK%', 'DBPM', 'DRB%', 'ORB%', 'DWS.1', 'OWS',

    # Advanced Metrics
    'VORP', 'PER', 'OnCourt', 'On-Off', 'NRtg/A',

    # Positional Information
    'PG%', 'SG%', 'SF%', 'PF%', 'C%',

    # Contract & Salary Data
    '2024-25_Player_Salary', '2025-26_Player_Salary', '2026-27_Player_Salary',
    '2027-28_Player_Salary', '2028-29_Player_Salary', 'Guaranteed_Player_Salary',

    # Team-Level Metrics
    'Team', 'W/L%', 'MOV', 'ORtg', 'DRtg', 'NRtg',

    # Team Payroll Data
    '2024-25_Payroll', '2025-26_Payroll', 'Value_per_Dollar', 'Cap_Impact_Score'
]


# Dropping rows with missing values in selected features
trade_data[features] = trade_data[features].fillna(0)

# Reduce dataset size for efficiency
trade_data_sample = trade_data.sample(n=min(500, len(trade_data)), random_state=42)

# Standardizing the features on a smaller dataset (excluding categorical 'Team' column)
features_numeric = [f for f in features if f != 'Team']
scaler = StandardScaler()
scaled_features_sample = scaler.fit_transform(trade_data_sample[features_numeric])

# Range of clusters to test
k_values = range(2, 11)

# Store clustering metrics
wcss = []  # Elbow method
silhouette_scores = []
calinski_scores = []
davies_bouldin_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(scaled_features_sample)
    
    # Compute evaluation metrics
    wcss.append(kmeans.inertia_)  # Sum of squared distances to closest centroid
    silhouette_scores.append(silhouette_score(scaled_features_sample, cluster_labels))
    calinski_scores.append(calinski_harabasz_score(scaled_features_sample, cluster_labels))
    davies_bouldin_scores.append(davies_bouldin_score(scaled_features_sample, cluster_labels))

# Create a figure with subplots to display all metrics side by side
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# Plot Elbow Method (WCSS)
axes[0].plot(k_values, wcss, marker='o', linestyle='-', color='b')
axes[0].set_xlabel('Number of Clusters (k)')
axes[0].set_ylabel('WCSS (Within-Cluster Sum of Squares)')
axes[0].set_title('Elbow Method for Optimal k')

# Plot Silhouette Score
axes[1].plot(k_values, silhouette_scores, marker='o', linestyle='-', color='g')
axes[1].set_xlabel('Number of Clusters (k)')
axes[1].set_ylabel('Silhouette Score')
axes[1].set_title('Silhouette Score for Different k Values')

# Plot Calinski-Harabasz Index
axes[2].plot(k_values, calinski_scores, marker='o', linestyle='-', color='r')
axes[2].set_xlabel('Number of Clusters (k)')
axes[2].set_ylabel('Calinski-Harabasz Index')
axes[2].set_title('Calinski-Harabasz Index for Different k Values')

# Plot Davies-Bouldin Index (lower is better)
axes[3].plot(k_values, davies_bouldin_scores, marker='o', linestyle='-', color='purple')
axes[3].set_xlabel('Number of Clusters (k)')
axes[3].set_ylabel('Davies-Bouldin Index (Lower is Better)')
axes[3].set_title('Davies-Bouldin Index for Different k Values')

# Display the combined plots
plt.tight_layout()
plt.show()

```

## 2. Choose Best Clustering Algorithm:

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from scipy.cluster.hierarchy import dendrogram, linkage

# Step 2: Choose the Best Clustering Algorithm

# **K-Means Clustering (Primary Method)**
best_k = 7  # Select k with the highest silhouette score
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(scaled_features_sample)

# **Hierarchical Clustering (Alternative Approach)**
hierarchical = AgglomerativeClustering(n_clusters=best_k, linkage='ward')
hierarchical_labels = hierarchical.fit_predict(scaled_features_sample)

# **Gaussian Mixture Models (If Needed for Flexible Clusters)**
gmm = GaussianMixture(n_components=best_k, random_state=42, covariance_type='full')
gmm_labels = gmm.fit_predict(scaled_features_sample)

# **Dendrogram for Hierarchical Clustering**
plt.figure(figsize=(10, 5))
linkage_matrix = linkage(scaled_features_sample, method='ward')
dendrogram(linkage_matrix, truncate_mode='level')  # Show the top 5 levels
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Player Index")
plt.ylabel("Distance")
plt.show()

# **Store Results in a DataFrame for Comparison**
trade_data_sample["KMeans_Cluster"] = kmeans_labels
trade_data_sample["Hierarchical_Cluster"] = hierarchical_labels
trade_data_sample["GMM_Cluster"] = gmm_labels

# Save the clustered results for download
clustered_file_path = "Stats/Clustered_Player_Segmentation.csv"
trade_data_sample.to_csv(clustered_file_path, index=False)

# Provide the download link
clustered_file_path
```
